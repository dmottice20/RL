```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Read in the data.
data = pd.read_csv('hw1problem8results.csv')
data.head()
```

```python
# Do Question 8 Part iv.
# Display optimal value fx. in a figure.
# On same figure, display value fx from early iteration.
part4data = data[['S', 'V_optimal', 'V_early_iteration']]
plt.plot(part4data['S'], part4data['V_optimal'], label='Optimal Policy from PI')
plt.plot(part4data['S'], part4data['V_early_iteration'], label='Early Iteration from PI')
plt.ylabel('Value Function, V(S)')
plt.xlabel('State (Age of Car), S')
plt.legend()
plt.show()
```

```python
# Do Question 8 Part v.
x = data['S']
y1 = data['pi_optimal']
y2 = data['pi_early_iteration']
plt.plot(x, y1, label='Optimal Policy from PI')
plt.plot(x, y2, label='Early Iteration from PI')
plt.ylim(0, 1600)
plt.ylabel('Decision function, d(S)')
plt.xlabel('State (Age of Car), S')
plt.annotate('{}'.format(y1.values[-1]), (x.values[-1], y1.values[-1]), xytext=(-10, 5), textcoords='offset points')
plt.annotate('{}'.format(y2.values[-1]), (x.values[-1], y2.values[-1]),  xytext=(-10, 5), textcoords='offset points')
plt.legend()
plt.show()
```
